### Introduction
This is a dataset based on the performance of students in an exam. The dataset consists of 1000 rows and 8 columns. The columns are;gender,race, lunch,parental level of education,test preparation course, math score,reading score and writing score. This analysis was conducted to find out what features directly affected the performance of students in an exam.

### Features
The key features of my dataset are math,reading and writing score, these contain the exam performance of each student. The main factors that may affect these features which are going to be analysed are;
###### parental level of education
###### test preparation course
###### gender

```python
#import the libraries needed for this analysis
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ex=pd.read_csv('exams.csv')
ex.head()
```

```python
ex.info()
```

```python
#renaming all columns to replace " " witb "_"
ex.rename(columns=lambda x:x.strip().replace(" ","_"),inplace=True)
```

```python
ex.head()
```

```python
#renaming columns race/ethnicity and parental level of education
ex.rename(columns={'race/ethnicity':'race','parental_level_of_education':'parent_qualification'},inplace=True)
```

```python
#checking for the distribution of male and female students
ex.gender.value_counts()
```

### How did students perform in their math exam?

```python
gender_perf_math=ex.groupby('gender')['math_score'].mean()
```

```python
plt.figure(figsize=[8,8])
plt.title('Math performance of students by gender',fontsize=16)
plt.xlabel('Gender',fontsize=12)
plt.ylabel('Average scores',fontsize=12)
gender_perf_math.plot(kind='bar', color='b')
plt.xticks(rotation=45);
```

Male students generally performed better in their math exam

### How did students perform in their reading exam?

```python
gender_perf_read=ex.groupby('gender')['reading_score'].mean()
```

```python
plt.figure(figsize=[8,8])
plt.title('Reading performance of students by gender',fontsize=16)
plt.xlabel('Gender',fontsize=12)
plt.ylabel('Average scores',fontsize=12)
gender_perf_read.plot(kind='bar',color='b')
plt.xticks(rotation=45);
```

From the visualization it is clearly shown that female students generally performed better in reading exams

### How did students perform in their writing exam?

```python
gender_perf_writing=ex.groupby('gender')['writing_score'].mean()
```

```python
plt.figure(figsize=[8,8])
plt.title('Writing performance of students by gender',fontsize=16)
plt.xlabel('Gender',fontsize=12)
plt.ylabel('Average scores',fontsize=12)
gender_perf_writing.plot(kind='bar',color='b')
plt.xticks(rotation=45);
```

it is clearly shown in the visualized data that female students generally performed better in writing exams

### Does parent level of education have an effect on students exam performance?

```python
#Grouping parents qualification based on math scores
math_perf=ex.groupby('parent_qualification')['math_score'].mean().sort_values(ascending=False)
```

```python
#visualizing the grouped data
plt.figure(figsize=[8,8]);
plt.title('students performance in math exam based on parents qualification',fontsize=16)
plt.xlabel('parent qualificaton',fontsize=12)
plt.ylabel('Average scores',fontsize=12)
math_perf.plot(kind='bar',color='b')
plt.xticks(rotation=45);
```

```python
#Grouping parents qualification based on reading scores
read_perf=ex.groupby('parent_qualification')['reading_score'].mean().sort_values(ascending=False)
```

```python
#visualizing the grouped data
plt.figure(figsize=[8,8]);
plt.title('students performance in reading exam based on parents qualification',fontsize=16)
plt.xlabel('parent qualificaton',fontsize=12)
plt.ylabel('Average scores',fontsize=12)
read_perf.plot(kind='bar',color='b')
plt.xticks(rotation=45);
```

```python
#Grouping parents qualification based on writing scores
writing_perf=ex.groupby('parent_qualification')['writing_score'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=[8,8]);
plt.title('students performance in writing exam based on parents qualification',fontsize=16)
plt.xlabel('parent qualificaton',fontsize=12)
plt.ylabel('Average scores',fontsize=12)
writing_perf.plot(kind='bar',color='b')
plt.xticks(rotation=45);
```

From the visualized exams data above, it is shown that students that have parents with a masters degree generally performed best, compared to students whose parents only have a high school certificate having the least performance in their exams.

### Is there a correlation between the math and reading scores of students?

```python
#a scatter plot to show the correlation between math and reading scores
plt.figure(figsize=[8,8])
sb.regplot(data=ex,x='math_score', y='reading_score', scatter_kws={'alpha':0.5})
plt.title('Relationship between math and reading score',fontsize=16);
```

The scatter plot shows a positive correlation between math and reading scores, indicating students that scored high in their math exam also scoring high marks in their reading exams and vice versa.

### How did students perform in their reading and writing exams?

```python
plt.figure(figsize=[8,8])
bin_x=np.arange(20,100+10,10)
bin_y=np.arange(20,100+10,10)
plt.hist2d(data=ex,x='reading_score',y='writing_score',cmin=0.5,cmap='viridis_r',bins=[bin_x,bin_y])
plt.title('Heatmap of students performance in reading and writing exams',fontsize=16)
plt.colorbar();
```

The heat map showed a bulk of the students scored between 60 and 80 in their reading and writing exams

### How did parents_qualification affect students' math exam performance

```python
plt.figure(figsize=[8,8])
color=sb.color_palette()[0]
sb.boxplot(data=ex,x='parent_qualification',y='math_score',color=color)
plt.title('Students performance in math exam based on parents qualification',fontsize=16)
plt.xticks(rotation=45);
```

The box plot showed majority of the top performing students with scores of at least 80 marks have parents with bachelor's or master's degree. students having parents with some high school certficate performed the poorest, with a majority of them scoring less than 70.

### Conclusion

Over the course of this analysis, parents educational qualification was seen to play a huge role in determining the scores of students, where parents with master's and bachelor's degrees produced students with the best results. Female students generally performed better in their reading and writing exam, while the male students performed better in their math exam. Also a strong postitive correlation was observed in reading and writing scores of students, meaning students that performed well in the reading exams also did well in their writing exams also.
